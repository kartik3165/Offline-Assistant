```python
from tkinter import *
import json
import os

def load_songs():
    if os.path.exists('songs.json'):  # Check if the JSON file exists
        with open('songs.json', 'r') as f:
            songs = json.load(f)
            for song in songs:
                song_listbox.insert(END, f"{song['name']} - {song['link']}")  # Add each song to the listbox

def add_song():
    song_name = song_name_entry.get()
    song_link = song_link_entry.get()
    
    if song_name and song_link:  # Check if both fields are filled
        # Add to Listbox
        song_listbox.insert(END, f"{song_name} - {song_link}")
        
        # Save to JSON
        save_song(song_name, song_link)
        
        # Clear the entry fields
        song_name_entry.delete(0, END)  
        song_link_entry.delete(0, END)  
    else:
        show_toast("Please enter both song name and song link.")

def save_song(name, link):
    # Load existing songs
    if os.path.exists('songs.json'):
        with open('songs.json', 'r') as f:
            songs = json.load(f)
    else:
        songs = []

    # Add new song
    songs.append({"name": name, "link": link})

    # Save back to JSON
    with open('songs.json', 'w') as f:
        json.dump(songs, f)

def show_toast(message):
    toast = Label(root, text=message, bg="black", fg="white", padx=10, pady=5)
    toast.place(relx=0.5, rely=0.5, anchor="center")
    root.after(2000, toast.destroy)

root = Tk()
root.geometry("800x500")
root.minsize(400, 300)
root.maxsize(1200, 800)

frame_top = Frame(root, bg="lightblue")
frame_bottom = Frame(root, bg="lightgreen")

frame_top.pack(fill="both", expand=True)
frame_bottom.pack(fill="both", expand=True)

# Add title label in the top frame
title_label = Label(frame_top, text="Song List", font=("Arial", 24), bg="lightblue")
title_label.pack(pady=10)

# Create a Listbox to display songs
song_listbox = Listbox(frame_bottom, bg="white", font=("Arial", 14))
song_listbox.pack(fill=BOTH, expand=True, padx=20, pady=10)

# Create entry fields and button to add a new song
add_song_frame = Frame(frame_bottom, bg="lightgreen")
add_song_frame.pack(pady=10)

song_name_label = Label(add_song_frame, text="Song Name:", bg="lightgreen")
song_name_label.grid(row=0, column=0, padx=5)

song_name_entry = Entry(add_song_frame, width=30)
song_name_entry.grid(row=0, column=1, padx=5)

song_link_label = Label(add_song_frame, text="Song Link:", bg="lightgreen")
song_link_label.grid(row=1, column=0, padx=5)

song_link_entry = Entry(add_song_frame, width=30)
song_link_entry.grid(row=1, column=1, padx=5)

add_song_button = Button(add_song_frame, text="Add Song", command=add_song)
add_song_button.grid(row=2, columnspan=2, pady=10)

load_songs()  # Load songs from the JSON file when the app starts

root.mainloop()
```

```python
from tkinter import *
from tkinter import ttk, messagebox
import json
import os

# Function to show toast message
def show_toast(message):
    toast = Label(root, text=message, bg="black", fg="white", padx=10, pady=5)
    toast.place(relx=0.5, rely=0.5, anchor="center")
    root.after(2000, toast.destroy)

# Functions for the menu actions
def update_action():
    show_toast("Update selected!")

def developer_action():
    show_toast("Developer selected!")

def about_us_action():
    show_toast("About Us selected!")

def toggle_menu():
    if menu_frame.winfo_ismapped():
        menu_frame.place_forget()
    else:
        menu_frame.place(relx=1.0, rely=0.1, anchor="ne")

# Function to switch between frames
def show_frame(frame):
    frame.tkraise()

# Functions for managing songs
def load_songs():
    for item in song_table.get_children():
        song_table.delete(item)
    if os.path.exists('songs.json'):
        with open('songs.json', 'r') as f:
            try:
                songs = json.load(f)
                for song_name, song_link in songs.items():
                    song_table.insert("", END, values=(song_name, song_link))
            except json.JSONDecodeError:
                show_toast("Error loading songs from JSON. File might be corrupted.")

def add_song():
    song_name = song_name_entry.get()
    song_link = song_link_entry.get()
    if song_name and song_link:
        song_table.insert("", END, values=(song_name, song_link))
        save_song(song_name, song_link)
        song_name_entry.delete(0, END)
        song_link_entry.delete(0, END)
    else:
        show_toast("Please enter both song name and song link.")

def save_song(name, link):
    if os.path.exists('songs.json'):
        with open('songs.json', 'r') as f:
            songs = json.load(f)
    else:
        songs = {}
    songs[name] = link
    with open('songs.json', 'w') as f:
        json.dump(songs, f, indent=4)

def delete_song(song_name):
    if os.path.exists('songs.json'):
        with open('songs.json', 'r') as f:
            songs = json.load(f)
        if song_name in songs:
            del songs[song_name]
        with open('songs.json', 'w') as f:
            json.dump(songs, f, indent=4)

def delete_button_click(event):
    item = song_table.identify_row(event.y)
    if item:
        item_values = song_table.item(item, "values")
        song_name = item_values[0]
        confirm = messagebox.askyesno("Confirm Delete", f"Are you sure you want to delete '{song_name}'?")
        if confirm:
            delete_song(song_name)
            song_table.delete(item)

# Create main window
root = Tk()
root.geometry("800x500")
root.minsize(400, 300)
root.maxsize(1200, 800)

# Create container for frames
container = Frame(root)
container.pack(fill="both", expand=True)

# Create two frames - one for the main interface and one for the song list interface
frame_main = Frame(container)
frame_song_list = Frame(container)

for frame in (frame_main, frame_song_list):
    frame.grid(row=0, column=0, sticky="nsew")

# Create the main interface in frame_main
def left_button_action():
    show_frame(frame_song_list)  # Switch to song list frame

def center_button_action():
    show_toast("Center frame button clicked!")

def right_button_action():
    show_toast("Right frame button clicked!")

def add_text_and_button(frame, label_text, button_text, button_action):
    inner_frame = Frame(frame, bg=frame.cget("bg"))
    inner_frame.pack(expand=True)
    text_label = Label(inner_frame, text=label_text, bg=inner_frame.cget("bg"), fg="white")
    text_label.pack(pady=10)
    action_button = Button(inner_frame, text=button_text, command=button_action)
    action_button.pack(pady=10)

# Create top and bottom frames in the main interface
frame_top = Frame(frame_main, bg="lightblue")
frame_bottom = Frame(frame_main, bg="lightgreen")

frame_top.pack(fill="both", expand=True, side=TOP)
frame_bottom.pack(fill="both", expand=True, side=TOP)

# Create frames inside the bottom frame
frame_left = Frame(frame_bottom, bg="red")
frame_center = Frame(frame_bottom, bg="yellow")
frame_right = Frame(frame_bottom, bg="blue")

frame_left.pack(side="left", fill="both", expand=True)
frame_center.pack(side="left", fill="both", expand=True)
frame_right.pack(side="left", fill="both", expand=True)

# Add text label and button to each frame with unique functionality
add_text_and_button(frame_left, "Left Frame", "Left Button", left_button_action)
add_text_and_button(frame_center, "Center Frame", "Center Button", center_button_action)
add_text_and_button(frame_right, "Right Frame", "Right Button", right_button_action)

# Create a canvas for the circular button
canvas = Canvas(frame_top, width=100, height=100, bg="lightblue", highlightthickness=0)
canvas.create_oval(10, 10, 90, 90, fill="blue", outline="")
canvas.create_text(50, 50, text="Click Me", fill="white", font=("Arial", 10, "bold"))
canvas.pack(expand=True)
canvas.bind("<Button-1>", lambda event: show_toast("Button is clicked!"))

# Create song management interface in frame_song_list
frame_top_song = Frame(frame_song_list, bg="lightblue")
frame_bottom_song = Frame(frame_song_list, bg="lightgreen")

frame_top_song.pack(fill="x", expand=False)
frame_bottom_song.pack(fill="both", expand=True)

title_label = Label(frame_top_song, text="Song List", font=("Arial", 24), bg="lightblue")
title_label.pack(pady=10)

back_button = Button(frame_top_song, text="Back", command=lambda: show_frame(frame_main))
back_button.pack(side=LEFT, padx=10)

# Create a Treeview to display songs
columns = ("Song Name", "Song Link")
song_table = ttk.Treeview(frame_bottom_song, columns=columns, show='headings', height=15)
song_table.heading("Song Name", text="Song Name")
song_table.heading("Song Link", text="Song Link")
song_table.pack(fill=BOTH, expand=True, padx=20, pady=10)

song_table.bind("<Double-1>", delete_button_click)

# Entry fields and button to add a new song
add_song_frame = Frame(frame_bottom_song, bg="lightgreen")
add_song_frame.pack(pady=10)

song_name_label = Label(add_song_frame, text="Song Name:", bg="lightgreen")
song_name_label.grid(row=0, column=0, padx=5, sticky=W)

song_name_entry = Entry(add_song_frame, width=30)
song_name_entry.grid(row=0, column=1, padx=5)

song_link_label = Label(add_song_frame, text="Song Link:", bg="lightgreen")
song_link_label.grid(row=1, column=0, padx=5, sticky=W)

song_link_entry = Entry(add_song_frame, width=30)
song_link_entry.grid(row=1, column=1, padx=5)

add_song_button = Button(add_song_frame, text="Add Song", command=add_song)
add_song_button.grid(row=2, columnspan=2, pady=10)

load_songs()

# Show the main frame initially
show_frame(frame_main)

root.mainloop()
```

```python
import json

def load_assistant_settings():
    with open("settings.json", "r") as file:
        return json.load(file)

def get_assistant_info():
    settings = load_assistant_settings()
    
    if not settings:
        return None, None
    
    # Handling any changes in the JSON file dynamically
    assistant_name = settings.get("assistant_name", "Unknown Assistant")
    auto_start = settings.get("auto_start", 0)

    return assistant_name, auto_start

name, auto_start_value = get_assistant_info()
print(name)
print(auto_start_value)
```
